CREATE SILVER LAYER PAYROLL TABLE

```
spark.sql("use silver_layer00")
spark.sql("""
          create table if not exists silver_payroll5(
              strstore integer,
              strTitle string,
              curWages double,
              last_updated timestamp
          )

          """)
```

```
#to get the last process in specified time stamp(last_updated from silver_payroll)
last_processed_df=spark.sql("select max(last_updated) as last_processed from silver_payroll  ")
last_processed_timestamp=last_processed_df.collect()[0]['last_processed']

if last_processed_timestamp is None:
      last_processed_timestamp = "2000-01-01T00:00:00.000+00:00"
```

```
#create a view a temporary payroll table created in the db
spark.sql(f"""
          create or replace temporary view bronze_incremental_pay
          as select 
          strstore,strTitle,curWages,CURRENT_TIMESTAMP() AS last_updated from 
          proj_bronze_layer00.bronze_payroll p where p.ingestion_time >'{last_processed_timestamp}'
        
          """)
```

```
%sql
select * from bronze_incremental_pay;
```

```
spark.sql("""
MERGE INTO silver_payroll5 target
using bronze_incremental_pay source
on target.strstore=source.strstore
when matched then 
    update set *
when not matched then
    insert  *          """) 
```

```
%sql
select *  from silver_payroll5;
```

CREATE SILVER LAYER SALES TABLE


```
spark.sql("use silver_layer00")
spark.sql("""
          create table if not exists silver_sales4(
              strstore integer,
              
              cur2014 double,
              cur2015 double,

              last_updated timestamp
          )

          """)
```

```
last_processed_df=spark.sql("select max(last_updated) as last_processed from silver_sales  ")
last_processed_timestamp=last_processed_df.collect()[0]['last_processed']

if last_processed_timestamp is None:
      last_processed_timestamp = "2000-01-01T00:00:00.000+00:00"
```

```
spark.sql(f"""
          create or replace temporary view bronze_incremental_sales
          as select strstore,
              cur2014,
              cur2015,CURRENT_TIMESTAMP() AS last_updated from 
          proj_bronze_layer00.bronze_sales s where s.ingestion_time >'{last_processed_timestamp}'
          """)
```

```
spark.sql("""
MERGE INTO silver_sales4 target
using bronze_incremental_sales source
on target.strstore=source.strstore
when matched then 
    update set *
when not matched then
    insert  *          """) 
```

```
%sql
select * from silver_sales4;
```

CREATE SILVER LAYER PROPERTIES TABLE 



```
spark.sql("use silver_layer00")
spark.sql("""
          create table if not exists silver_prop(
              strstore integer,
              strstate string,
              curtotalproperty double,
              curstructures double,
              curequipment double,

              last_updated timestamp)

          """)
```

```
last_processed_df=spark.sql("select max(last_updated) as last_processed from silver_prop ")
last_processed_timestamp=last_processed_df.collect()[0]['last_processed']

if last_processed_timestamp is None:
      last_processed_timestamp = "2000-01-01T00:00:00.000+00:00"
```

```
spark.sql(f"""
          create or replace temporary view bronze_incremental_prop
          as select strstore,
          strstate,
          curtotalproperty,
          curstructures,
          curequipment,
          CURRENT_TIMESTAMP() AS last_updated  from 
          proj_bronze_layer00.bronze_property s where s.ingestion_time >'{last_processed_timestamp}'
          """)
```

```
spark.sql("""
MERGE INTO silver_prop target
using bronze_incremental_prop source
on target.strstore=source.strstore
when matched then 
    update set *
when not matched then
    insert  *          """) 
```

```
%sql
select * from silver_prop;
```

Question 1: Which store manager(s) make more than $100,000?

```
spark.sql("""
          create or replace temporary view silver_Managers_above as select * from silver_payroll5 where strTitle="Manager" and curWages>100000;
          """)
```

```
%sql
select * from silver_Managers_above;
```

Question 2: What is the total payroll for each position where total wages exceed $700,000?

```
spark.sql("""
          create or replace temporary view silver_total_position_payroll as select round(sum(curWages),2) sum,strTitle from silver_payroll5
            group by strTitle
            having sum(curWages)>700000;
            """)
```

```
%sql
select * from silver_total_position_payroll;
```

Question 3: The company wants to give a 3% wage increase to managers whose 2015 store sales is bigger than their 2014 store sales.

```
#create a view a temporary table created in the db
spark.sql(f"""
          create or replace temporary view silver_manager_pay_increase
          as select  pay.strstore,strTitle,curWages,round((curwages*0.03),2)as CurWagesInc ,round(CurWagesInc+curWages) as CurWagesTotal  from 
           silver_sales4 sales join silver_payroll5 pay 
           where strTitle=="Manager" 
           group by  pay.strstore,strTitle,curWages

           having (sum(cur2014)<sum(cur2015))
           order by strTitle;      
          """)
```

```
%sql
       select * from  silver_manager_pay_increase
       order by CurWagesTotal desc;
```

   Question 4: Feasibility of a 3% Payroll Increase (Company-Wide)


```
%sql
select strstore,strTitle,sum(curWages), round((sum(CurWagesTotal)-sum(CurWages)),2) as difference_per_store  from silver_manager_pay_increase
group by strstore, strTitle
order by difference_per_store desc;
```

Question 2: Which store(s) or region(s) have the largest gap between 2014 and 2015 sales?


```
%sql
select p.strstore,strstate, cur2015-cur2014 as 2015_2014_gap from silver_sales4 p join silver_prop pp on p.strstore=pp.strstore order by 
2015_2014_gap desc;
```
